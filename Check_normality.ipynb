```python
import pandas as pd
import numpy as np
df= pd.read_csv('data.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
plt.figure(figsize=(20,16))

sns.histplot(df[' Blood Pressure Before (mmHg)'] , kde=True)
plt.show()
```

```python
from scipy.stats import shapiro
stat, p = shapiro(df[' Blood Pressure Before (mmHg)'])
print(p)
```

```python
As p value is not greater than 0.05 means data is not gaussian .
```

```python
plt.figure(figsize=(20,16))

sns.histplot(df[' Blood Pressure After (mmHg)'] , kde=True)
plt.show()
```

```python
from scipy.stats import shapiro
stat, p = shapiro(df[' Blood Pressure After (mmHg)'])
print(p)
```

```python
As p value is less than 0.05 means the data is not normally distributed .
```
