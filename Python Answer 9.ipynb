```python
● Draw plots to show the distribution between city and electric vehicle type
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

countt= df['City'].value_counts()
countt

```

```python
plt.figure(figsize=(10, 5))
plt.bar(countt.index, countt.values)
plt.show()
```

```python
ev_type_counts = df['Electric Vehicle Type'].value_counts()

# Plot 1: Distribution of cities
plt.figure(figsize=(10, 5))
plt.bar(ev_type_counts.index, ev_type_counts.values)
plt.xlabel('vehicle type')
plt.ylabel('Count')
plt.title('Distribution of vehicle type')
plt.show()
```
