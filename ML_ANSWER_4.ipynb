```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder() 

df['VisitorType']=labelencoder.fit_transform(df['VisitorType'])
df['VisitorType']=df['VisitorType'].astype('int')

df['Month']=labelencoder.fit_transform(df['Month'])
df['Month']=df['Month'].astype('int')

df['Weekend']=labelencoder.fit_transform(df['Weekend'])
df['Weekend']=df['Weekend'].astype('int')

df['Revenue']=labelencoder.fit_transform(df['Revenue'])
df['Revenue']=df['Revenue'].astype('int')
```

```python
df.head()
```

```python
X = df.drop(labels=['Revenue'],axis=1)
Y = df.Revenue
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=30)

scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test =scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import xgboost as xg
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV 
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV 
```

```python
df_random= RandomForestClassifier(n_estimators=5)
```

```python
df_random.fit(X_train,y_train)
```

```python
df_random.score(X_train,y_train)
```

```python
grid_param={
    'n_estimators':[5,10,15,20,25,50],
    'criterion':['gini', 'entropy'],
    'max_depth':range(20),
    'min_samples_leaf':range(10)
}
grid_search_cv = GridSearchCV(param_grid= grid_param , estimator= df_random, cv=10 , n_jobs=10, verbose=1)
grid_search_cv.fit(X_train,y_train)
```

```python
grid_search_cv.best_params_
```

```python
df_best = RandomForestClassifier(criterion= 'entropy',
 max_depth=9,
 min_samples_leaf=6,
 n_estimators=50)
```

```python
df_best.fit(X_train,y_train)
```

```python
y_pred = df_best.predict(X_test)   ##revenue sales predicted
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
