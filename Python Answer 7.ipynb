```python
import pandas as pd
import numpy as np

df= pd.read_csv('gov.csv')
```

```python
df.drop(labels="Unnamed: 0", axis=1, inplace= True)
```

```python
df['year']= np.where(df['year'].isnull() , "1933-01-01T00:00:00.000", df['year']) #fill missing values
```

```python
df['Year']=  df['year'].str.split('-').str[0] ##extract exact year from year
```

```python
df['Year']=df['Year'].astype('int')
```

```python
df['Year'].value_counts()
```

```python
● Get all the Earth meteorites that fell before the year 2000
```

```python
df[df['Year']<2000]  #data for meteorities that fell before year 2000
```

```python
#how many meteorites co-ordinates who fell before the year 2000
df[df['Year']<2000].shape[0]
```

```python
#Get all the earth meteorites co-ordinates who fell before the year 1970
df[df['Year']<1970]
```

```python
● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more
than 10000kg
```

```python
df_mass= df[df['mass']> 10000]
```
